```python
# determining if outliers are present
import seaborn as sns
sns.boxplot(df["Value"])
```

From the boxplot above, the black circular points indicate the presence of extreme values in Value

```python
# creating column Value_new and replacing missing values with the median
df["Value_new"] = df["Value"]
df["Value_new"] = df["Value_new"].fillna(df["Value_new"].median())
```

```python
sns.boxplot(df["International Reputation"])
```

From the boxplot above, there are outliers at 2.0, 3.0, 4.0 & 5.0

```python

df["International Reputation_new"] = df["International Reputation"]
df["International Reputation_new"] = df["International Reputation_new"].fillna(df["International Reputation_new"].median())
```

```python
sns.boxplot(df["Skill Moves"])
```

```python

df["Skill Moves_new"] = df["Skill Moves"]
df["Skill Moves_new"] = df["Skill Moves_new"].fillna(df["Skill Moves_new"].median())
```

**For catergorical features, we will impute with "Missing"**

```python
df["Club_new"] = df["Club"]
df["Club_new"] = df["Club_new"].fillna("Missing")

df["Contract Valid Until_new"] = df["Contract Valid Until"]
df["Contract Valid Until_new"] = df["Contract Valid Until_new"].fillna("Missing")
```

Dropping columns with null values

```python
df = df.drop(["Club","Value","International Reputation","Skill Moves","Contract Valid Until"], axis=1)
```

**Let's find the players that have an overall rating more than 90 and skill move greater than 3.**

```python

df1 = df.groupby(['Name'])['Overall'].mean().sort_values(ascending=False).reset_index()


df1.rename(columns={'Overall': 'Overall_avg'}, inplace=True)


df1 = df1[df1['Overall_avg']>90]


df1
```

```python

df2 = df.groupby(['Name'])['Skill Moves_new'].mean().sort_values(ascending=False).reset_index()


df2.rename(columns={'Skill Moves_new': 'Skill Moves_new_avg'}, inplace=True)


df2 = df2[df2['Skill Moves_new_avg']>3]


df2
```

```python
df3 = pd.merge(df1,df2,how='inner',on='Name')
df3
```
